```python
!wget https://github.com/ultralytics/assets/releases/download/v8.3.0/yolo11m.pt -O /kaggle/working/yolov11m.pt

# Load YOLO model
model = YOLO('yolov11m.pt')  # Make sure you have YOLOv11m weights in the working directory

# Train the model with the current hyperparameters
results = model.train(
        data='data.yaml',
        epochs=100,
        imgsz=640,
        seed=42,
        batch=16,
        augment=True,
        workers=4,
        lr0=0.01,
        lrf=0.01,
        momentum=0.9,
        weight_decay=0.0005,
        optimizer='SGD',
        patience=10,
        project='Wrist_m_640_32_SGD',  # Define custom project path
        warmup_epochs=3
    )
```
